## **HARJOITUS 2:** 










Etsi allaolevien datapisteiden $x^{(i)}=[x_1^{(i)},x_2^{(i)}],\,y^{(i)}=0,1$, 
perusteella kertoimet $w_1,w_2,w_3,w_4,w_5$ ja $b$ niin, että piste $x=[x_1,x_2]$ kuuluu 

luokkaan $y=1$, jos
$z=w_1x_1^2+w_2x_1x_2+w_3x_2^2+w_4x_1+w_5x_2+b> 0$

 luokkaan $y=0$, jos $z<0$


käyttämällä logistista regressiota datapisteille


 $$[(x^{(i)}_1)^2,x^{(i)}_1x^{(i)}_2,(x^{(i)}_2)^2,x^{(i)}_1,x^{(i)}_2 ],\,y^{(i)}$$


```python
#datapisteet
np.random.seed(0)

r1=np.random.rand(100)
th1=2*np.pi*np.random.rand(100)
x1pun=r1*np.cos(th1)
x2pun=1.5*r1*np.sin(th1)

r2=2+np.random.rand(100)
th2=2*np.pi*np.random.rand(100)
x1sin=r2*np.cos(th2)
x2sin=1.5*r2*np.sin(th2)


plt.figure(figsize=(6,6))
plt.plot(x1pun,x2pun,'r.',markersize=12,label='y=1')
plt.plot(x1sin,x2sin,'b.',markersize=12,label='y=0')
plt.grid()
plt.axis('equal')
plt.legend(fontsize=14)
plt.xlabel('x1',fontsize=12)
plt.ylabel('x2 ',rotation=0,fontsize=12)
plt.show()
```
