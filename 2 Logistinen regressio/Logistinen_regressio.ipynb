## **Logistinen regressio (Logistic regression)** 

[DeepLearningAI: Neural Networks and Deep Learning, 7-24](https://www.youtube.com/playlist?list=PLkDaE6sCZn6Ec-XTbcX1uRg2_u4xOEky0)

Tavoitteena on etsiä laskukaava pisteiden $x=[x_1,x_2,...,x_n]$ luokitteluun muuttujien $x_1,x_2,\dots,x_n$ avulla.

Laskukaavan kertoimet  määrätään datapisteiden perusteella minimoimalla kustannusfunktio.

2 luokkaa: Binary classification

3,4,5... luokkaa: Multiclass classification

## **2 muuttujaa, 2 luokkaa** 



Datapisteet 
$x^{(i)}=[x^{(i)}_1,x^{(i)}_2]$
 ja niiden luokat 
 $y^{(i)}=0$ tai 1, $\quad i=1,2,\dots ,m$

 Etsitään niiden perusteella kertoimet $w_1,w_2$ ja $b$ niin, että eri luokkien pisteet ovat "mahdollisimman hyvin" eri puolella suoraa (decision boundary)
 
 
$$
w_1x_1+w_2x_2+b= 0
$$ 

image.png

Pisteen  $x=[x_1,x_2]$ luokittelu: 

Jos 
$
z=w_1x_1+w_2x_2+b>  0
$
eli piste on suoran "yläpuolella", niin $x$ kuuluu luokkaan $y=1$



Jos
$
z=w_1x_1+w_2x_2+b<  0
$ 
eli piste on suoran "alapuolella", niin $x$ kuuluu luokkaan $y=0$. 

(Suoralla olevat pisteet jompaan kumpaan luokkaan)

**Geometria:** suora 
$
w_1x_1+w_2x_2+b=0
$
 on kohtisuorassa vektoria $w=[w_1,w_2]$ vastaan, ja kulkee pisteen

$$
P=-\,\frac{b}{|w|}\cdot \left[\frac{w_1}{|w|},\frac{w_2}{|w|}\right]
$$

kautta, missä $|w|=\sqrt{w_1^2+w_2^2}$ on $w$:n pituus 


(eli $OP$:n pituus on $|b|/|w|$ ja suunta $-w$, jos $b>0$ ja $w$, jos $b<0$). 


$$
\begin{array}{l}\\\\
\end{array}
$$




$z=w_1x_1+w_2x_2+b>0$
suoran $w$:n puolella ja $z$:n arvo  kasvaa, kun liikutaan poispäin suorasta $w$:n suuntaan
 
 $$
\begin{array}{l}\\
\end{array}
$$
	
	

	
$z=w_1x_1+w_2x_2+b<0$
suoran $-w$:n puolella ja $z$:n arvo pienenee, kun liikutaan poispäin suorasta  $-w$:n  suuntaan


$$
\begin{array}{l}\\\\
\end{array}
$$

suorien $w_1x_1+w_2x_2+b=\pm d$ etäisyys suorasta $w_1x_1+w_2x_2+b=0$ on $d/|w|$



image.png

```python
#suorat w1*x1+w2*x2+b=0 ja w1*x1+w2*x2+b=d
w1=-1
w2=2
b=1
d=3

nw=np.sqrt(w1**2+w2**2) #|w|

#suoran w1*x1+w2*x2+b=0 piste P
px1=-b/nw*w1/nw
px2=-b/nw*w2/nw
sk=np.arctan2(w2,w1) #w:n suuntakulma
L=4 #suoran pituus = 2L
#suoran päätepisteet
p1x1=px1+L*np.cos(sk+np.pi/2)
p1x2=px2+L*np.sin(sk+np.pi/2)
p2x1=px1+L*np.cos(sk-np.pi/2)
p2x2=px2+L*np.sin(sk-np.pi/2)

#suoran w1*x1+w2*x2+b=d eli w1*x1+w2*x2+b-d=0 piste Q
qx1=-(b-d)/nw*w1/nw
qx2=-(b-d)/nw*w2/nw
#suoran päätepisteet
q1x1=qx1+L*np.cos(sk+np.pi/2)
q1x2=qx2+L*np.sin(sk+np.pi/2)
q2x1=qx1+L*np.cos(sk-np.pi/2)
q2x2=qx2+L*np.sin(sk-np.pi/2)

plt.figure(figsize=(5,5))
plt.plot([0,w1],[0,w2],'k',linewidth=2,label='w')
plt.plot([p1x1,p2x1],[p1x2,p2x2],'g',label='$w_1x_1+w_2x_2+b=0$')
plt.plot(px1,px2,'go',label='P')
plt.plot([q1x1,q2x1],[q1x2,q2x2],'b',label='$w_1x_1+w_2x_2+b=d$')
plt.grid()
plt.axis('equal')
plt.legend(fontsize=12)
plt.title('$w_1$ = '+str(w1)+', $w_2$ = '+str(w2)+', b = '+str(b)+', d = '+str(d))
plt.show()
```

Suoran kertoimet $w_1,w_2$ ja $b$ määrätään datapisteiden perusteella niin, että luokan 1 pisteet ovat mahdollisimman kaukana suorasta $w$:n suuntaan ja luokan 0 pisteet $-w$:n suuntaan, eli niin, että 

$$z^{(i)} = w_1x^{(i)}_1+w_2x^{(i)}_2+b $$
 on 

$$\begin{array}{l}
\textrm{suuri positiivinen, jos } y^{(i)}=1  \\
 \textrm{suuri negatiivinen, jos } y^{(i)}=0 \end{array}
$$

eli niin, että 

$$
  a^{(i)}=\frac{1}{1+\exp(-z^{(i)})}
	\approx \left\{
	\begin{array}{l}
	1,\textrm{ jos } y^{(i)}=1\\
	0,\textrm{ jos } y^{(i)}=0\\
	\end{array}\right.
$$

$$
\begin{array}{l}\\\end{array}
$$

missä $\exp(-z)=e^{-z}$, $e\approx 2.718...$, on eksponenttifunktio ja

$$
\sigma(z)=\frac{1}{1+\exp(-z)}
$$
on  **sigmoid**-funktio 


```python
#sigmoid
z=np.linspace(-6,6,100)
s=1/(1+np.exp(-z))

plt.figure(figsize=(6,4))
plt.plot(z,s)
plt.grid()
plt.xlabel('z')
plt.ylabel(r'$\sigma(z)$',rotation=0)
plt.show()
```

Tulkinta: Piste $x=[x_1,x_2]$ 
kuuluu 
luokkaan 1 todennäköisyydellä 
$a=\sigma(z)$
ja luokkaan 0 todennäköisyydellä $1-a$
, missä  $z=w_1x_1+w_2x_2+b$

Datapisteen kustannus 
$$\begin{array}{rl}
C^{(i)}&=  \left\{\begin{array}{l}
-\log(a^{(i)}), \textrm{ jos }y^{(i)}=1\\\\
-\log(1-a^{(i)}), \textrm{ jos }y^{(i)}=0
\end{array}
\right. \\\\
&=-y^{(i)}\log(a^{(i)})-(1-y^{(i)})\log(1-a^{(i)})
\end{array}
$$


missä $\log(...)$ on (luonnollinen) logaritmi.


$$
\begin{array}{l}\\
\end{array}
$$


Jos pisteen luokka $y^{(i)}=1$, niin
$$C^{(i)}=-\log(a^{(i)})=
\left\{
\begin{array}{rl}
\approx 0,&\textrm{ jos }a^{(i)}\approx 1\\
\textrm{suuri},&\textrm{  jos } a^{(i)}\approx 0
\end{array}\right.
$$

ja jos luokka $y^{(i)}=0$,  niin 
$$C^{(i)}=-\log(1-a^{(i)}) =
\left\{
\begin{array}{rl}
\approx 0,&\textrm{ jos }a^{(i)}\approx 0\\
\textrm{suuri},&\textrm{  jos } a^{(i)}\approx 1
\end{array}\right. 
$$ 


```python
#-log(ai) ja -log(1-ai):n kuvaajat, un ai=0...1
a=np.linspace(0.001,0.999,100)
C1=-np.log(a)
C0=-np.log(1-a)
plt.figure(figsize=(5,5))
plt.plot(a,C1,'r',label='- log(a)')
plt.plot(a,C0,'b',label='- log(1-a)')
plt.grid()
plt.xlabel('a')
plt.legend()
plt.show()
```

Kertoimet $w_1,w_2$ ja $b$ löydetään minimoimalla datapisteiden kustannusten keskiarvo eli kustannusfunktio (**binary cross-entropy**) 

$$
C=\displaystyle{\,-\,\frac{1}{m}  \,\sum_{i=1}^m y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})}$$ 


```python
#luodaan datapisteet ja niiden luokat
k=100
m=2*k #datapisteiden määrä
P1=np.array([4,2])
P2=np.array([-1,-1])
r=1
np.random.seed(0)
xpun=P1+r*np.random.randn(k,2)#luokka y=1
xsin=P2+r*np.random.randn(k,2)#luokka y=0 
x=np.vstack((xpun,xsin)) #(m,2)-matriisi, koordinaatit vaakariveinä

x1=x[:,0]
x2=x[:,1]

ypun=np.ones(k)
ysin=np.zeros(k)
y=np.hstack((ypun,ysin)) #luokat

#suora w1*x1+w2*x2+b=0
w1=2
w2=1
b=-2

nw=np.sqrt(w1**2+w2**2) #|w|
#suoran piste P
px1=-b/nw*w1/nw
px2=-b/nw*w2/nw
sk=np.arctan2(w2,w1) #w:n suuntakulma
L=6 #suoran pituus = 2L
#suoran päätepisteet
p1x1=px1+L*np.cos(sk+np.pi/2)
p1x2=px2+L*np.sin(sk+np.pi/2)
p2x1=px1+L*np.cos(sk-np.pi/2)
p2x2=px2+L*np.sin(sk-np.pi/2)
#datapisteiden kustannukset

z=w1*x1+w2*x2+b
a=1/(1+np.exp(-z))
Ci=-y*np.log(a+1e-10)-(1-y)*np.log(1-a+1e-10)#y=1->log(a), y=0->log(1-a)
#1e-10 jotta ei tulisi log(0):aa 

#kustannusfunktio
C=1/m*np.sum(Ci) 

plt.figure(figsize=(5,5))
plt.plot(xpun[:,0],xpun[:,1],'r.',label='y=1',markersize=10)
plt.plot(xsin[:,0],xsin[:,1],'b.',label='y=0',markersize=10)
plt.plot([p1x1,p2x1],[p1x2,p2x2],'k')
plt.grid()
plt.axis('equal')
plt.xlabel('x1',fontsize=14)
plt.ylabel('x2',fontsize=14,rotation=0)
plt.title('w1 = '+str(w1)+', w2 = '+str(w2)+', b = '+str(b)+',  C = {:5f}'.format(C))
plt.show()
```

Gradient descent: valitaan kertoimille $w_1,w_2$ ja $b$ alkuarvot ja päivitetään kertoimia

$$
\begin{array}{l}
\displaystyle{w_j=w_j-\alpha\cdot \frac{\partial C}{\partial w_j},\quad j=1,2}\\\\
\displaystyle{b=b-\alpha\cdot \frac{\partial C}{\partial b}}
\end{array}
$$







missä $\alpha>0$ on oppimisnopeus (learning rate)
,
ja osittaisderivaatat

$$\begin{array}{l}
\displaystyle{\frac{\partial C}{\partial w_j}=\frac{1}{m} \,\sum_{i=1}^m  (a^{(i)}-y^{(i)}) x^{(i)}_j},\quad j=1,2\\\\
\displaystyle{\frac{\partial C}{\partial b}=\frac{1}{m} \,\sum_{i=1}^m  (a^{(i)}-y^{(i)})}
\end{array}
$$

```python
#gradient descent 
N=1000 #kierrosmäärä
#kertoimien w1,w2 ja b alkuarvot
w1=np.random.randn()
w2=np.random.randn()
b=np.random.randn()
#oppimisnopeus
alfa=2
#vektorit, joihin kerätään kierrosten tulokset
C=np.zeros(N)
W1=np.zeros(N)
W2=np.zeros(N)
B=np.zeros(N)


for k in range(N): 
    z=w1*x1+w2*x2+b
    a=1/(1+np.exp(-z))
    Ci=-y*np.log(a+1e-10)-(1-y)*np.log(1-a+1e-10)#y=1->log(a), y=0->log(1-a)
    #1e-10 jotta ei tulisi log(0):aa 
    
    #kustannusfunktio
    C[k]=1/m*np.sum(Ci)

    #osittaisderivaatat
    dw1=1/m*np.sum((a-y)*x1)
    dw2=1/m*np.sum((a-y)*x2)
    db=1/m*np.sum(a-y)

    #kertoimien päivitys
    w1=w1-alfa*dw1
    w2=w2-alfa*dw2
    b=b-alfa*db
    
    W1[k]=w1
    W2[k]=w2
    B[k]=b
    
    

print(C[N-1])
print(w1)
print(w2)
print(b)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(C)
plt.grid()
plt.xlabel('kierros')
plt.ylabel('kustannus C')
   
plt.subplot(1,2,2)
plt.plot(W1,label='$w_1$')
plt.plot(W2,label='$w_2$')
plt.plot(B,label='$b$')
plt.grid()
plt.xlabel('kierros')
plt.legend()
plt.show()
```

```python
#kustannusfunktion minimoiva suora w1*x1+w2*x2+b=0
#ja syntyvä tason pisteiden luokittelu

#muodostetaan NxN grid-pistettä
N=100
xx1=np.linspace(-5,7,N)
xx2=np.linspace(-6,6,N)
X1,X2=np.meshgrid(xx1,xx2)#x1x2-parit, NxN-matriiseja

Z=w1*X1+w2*X2+b #vastaavat z:n arvot, NxN

#luokan 1 grid-pisteet
X1pun=X1[Z>=0] 
X2pun=X2[Z>=0]
#luokan 0 grid-pisteet
X1sin=X1[Z<0]
X2sin=X2[Z<0]

plt.figure(figsize=(5,5))
plt.plot(xpun[:,0],xpun[:,1],'ro',label='$y^{(i)}=1$',markersize=6)
plt.plot(xsin[:,0],xsin[:,1],'bo',label='$y^{(i)}=0$',markersize=6)
plt.plot(X1pun,X2pun,'r.',markersize=1)
plt.plot(X1sin,X2sin,'b.',markersize=1)
plt.contour(X1,X2,Z,levels=[0],colors=['k'],linewidths=[4]) #suora Z=0
plt.grid()
plt.axis('equal')
plt.title('w1 = {:.3f}'.format(w1)+', w2 = {:.3f}'.format(w2)+
          ', b = {:.3f}'.format(b),fontsize=14)
plt.xlabel('$x_1$',fontsize=14)
plt.ylabel('$x_2$',fontsize=14,rotation=0)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

Ongelma: kustannusfunktion arvo saadaan likimain nollaksi käyttäamällä mitä tahansa luokat erottavaa suoraa 
(vaikka se ei olisi "geometrisesti paras"): 

$$
w_1x_1+w_2x_2+b=0 \quad \textrm{ja}\quad  k w_1x_1+k w_2\,x_2+k b=0  
$$


ovat sama suora, ja 
jos luokkien 0 ja 1  pisteet ovat sen eri puolilla eli

$$\begin{array}{l}
z^{(i)}=w_1x_1+w_2x_2+b>0,\textrm{ kun } y^{(i)}=1\\\\
z^{(i)}=w_1x_1+w_2x_2+b<0,\textrm{ kun } y^{(i)}=0\end{array}
$$



ja jos $z^{(i)}\to k\,z^{(i)}$, niin 

$$
a^{(i)}= \frac{1}{1+\exp(-k\,z^{(i)})}\to 
\left\{ 
\begin{array}{l}
1,\textrm{ kun }y^{(i)}=1\\
0,\textrm{ kun }y^{(i)}=0
\end{array}\right.
$$



 kun $k$ kasvaa, eli kustannusfunktio $C\to 0$.



```python
#suora w1*x1+w2*x2+b=0
w1=2
w2=1
b=-2

k=10
w1*=k
w2*=k
b*=k

nw=np.sqrt(w1**2+w2**2) #|w|
#suoran piste P
px1=-b/nw*w1/nw
px2=-b/nw*w2/nw
sk=np.arctan2(w2,w1) #w:n suuntakulma
L=6 #suoran pituus = 2L
#suoran päätepisteet
p1x1=px1+L*np.cos(sk+np.pi/2)
p1x2=px2+L*np.sin(sk+np.pi/2)
p2x1=px1+L*np.cos(sk-np.pi/2)
p2x2=px2+L*np.sin(sk-np.pi/2)
#datapisteiden kustannukset

z=w1*x1+w2*x2+b
a=1/(1+np.exp(-z))
Ci=-y*np.log(a+1e-10)-(1-y)*np.log(1-a+1e-10)#y=1->log(a), y=0->log(1-a)
#1e-10 jotta ei tulisi log(0):aa 

#kustannusfunktio
C=1/m*np.sum(Ci) 

plt.figure(figsize=(5,5))
plt.plot(xpun[:,0],xpun[:,1],'r.',label='y=1',markersize=10)
plt.plot(xsin[:,0],xsin[:,1],'b.',label='y=0',markersize=10)
plt.plot([p1x1,p2x1],[p1x2,p2x2],'k')
plt.grid()
plt.axis('equal')
plt.xlabel('x1',fontsize=14)
plt.ylabel('x2',fontsize=14,rotation=0)
plt.title('w1 = '+str(w1)+', w2 = '+str(w2)+', b = '+str(b)+',  C = {:5f}'.format(C))
plt.show()
```

Ongelma voidaan korjata käayttämällä ns. **regularisoitua** kustannusfunktiota

$$\begin{array}{rl}
C=&\displaystyle{-\,\frac{1}{m}  \,\sum_{i=1}^m \log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})}
+\,\displaystyle{\frac{1}{2m}\left(w_1^2+w_2^2\right)}
\end{array}
$$


joka "sakottaa"  suurista kertoimien $w_1$ ja $w_2$ arvoista.


Osittaisderivaatat
$$\begin{array}{l}
\displaystyle{\frac{\partial C}{\partial w_j}=\frac{1}{m} \,\sum_{i=1}^m  (a^{(i)}-y^{(i)}) x^{(i)}_j}+\frac{1}{m}\,w_j,\quad j=1,2\\\\
\displaystyle{\frac{\partial C}{\partial b}=\frac{1}{m} \,\sum_{i=1}^m  (a^{(i)}-y^{(i)})}
\end{array}
$$

```python
#gradient descent 
N=1000 #kierrosmäärä
#kertoimien w1,w2 ja b alkuarvot
w1=np.random.randn()
w2=np.random.randn()
b=np.random.randn()
#oppimisnopeus
alfa=2
#vektorit, joihin kerätään kierrosten tulokset
C=np.zeros(N)
W1=np.zeros(N)
W2=np.zeros(N)
B=np.zeros(N)


for k in range(N): 
    z=w1*x1+w2*x2+b
    a=1/(1+np.exp(-z))
    Ci=-y*np.log(a+1e-10)-(1-y)*np.log(1-a+1e-10)#y=1->log(a), y=0->log(1-a)
    #1e-10 jotta ei tulisi log(0):aa 
    
    #kustannusfunktio
    C[k]=1/m*np.sum(Ci)+1/(2*m)*(w1**2+w2**2)

    #osittaisderivaatat
    dw1=1/m*np.sum((a-y)*x1)+1/m*w1
    dw2=1/m*np.sum((a-y)*x2)+1/m*w2
    db=1/m*np.sum(a-y)

    #kertoimien päivitys
    w1=w1-alfa*dw1
    w2=w2-alfa*dw2
    b=b-alfa*db
    
    W1[k]=w1
    W2[k]=w2
    B[k]=b
    
    

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(C)
plt.grid()
plt.xlabel('kierros')
plt.ylabel('kustannus C')
   
plt.subplot(1,2,2)
plt.plot(W1,label='$w_1$')
plt.plot(W2,label='$w_2$')
plt.plot(B,label='$b$')
plt.grid()
plt.xlabel('kierros')
plt.legend()
plt.show()
```

```python
#regularisoideun kustannusfunktion minimoiva suora w1*x1+w2*x2+b=0
#ja syntyvä tason pisteiden luokittelu

#muodostetaan NxN grid-pistettä
N=100
xx1=np.linspace(-5,7,N)
xx2=np.linspace(-6,6,N)
X1,X2=np.meshgrid(xx1,xx2)#x1x2-parit, NxN-matriiseja

Z=w1*X1+w2*X2+b #vastaavat z:n arvot, NxN

#luokan 1 grid-pisteet
X1pun=X1[Z>=0] 
X2pun=X2[Z>=0]
#luokan 0 grid-pisteet
X1sin=X1[Z<0]
X2sin=X2[Z<0]

plt.figure(figsize=(5,5))
plt.plot(xpun[:,0],xpun[:,1],'ro',label='$y^{(i)}=1$',markersize=6)
plt.plot(xsin[:,0],xsin[:,1],'bo',label='$y^{(i)}=0$',markersize=6)
plt.plot(X1pun,X2pun,'r.',markersize=1)
plt.plot(X1sin,X2sin,'b.',markersize=1)
plt.contour(X1,X2,Z,levels=[0],colors=['k'],linewidths=[4]) #suora Z=0
plt.grid()
plt.axis('equal')
plt.title('w1 = {:.3f}'.format(w1)+', w2 = {:.3f}'.format(w2)+
          ', b = {:.3f}'.format(b),fontsize=14)
plt.xlabel('$x_1$',fontsize=14)
plt.ylabel('$x_2$',fontsize=14,rotation=0)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

##  **$n$ muuttujaa, 2 luokkaa**  






Datapisteet
 $x^{(i)}=[x^{(i)}_1,x^{(i)}_2,\dots,x^{(i)}_n]$
  ja niiden luokat $y^{(i)}=0$ tai 1, $i=1,2,\dots ,m$. 


Etsitään  kertoimet 
$w=[w_1,w_2,\dots,w_n]$ ja $b$ niin, että eri luokkien pisteet ovat (mahdollisimman hyvin) eri puolilla 
$w$:tä vastaan kohtisuoraa tasoa

$$
w_1x_1+w_2x_2+\dots+w_nx_n+b=0
$$

image.png

eli niin, että

$$
z^{(i)}=w_1x^{(i)}_1+w_2x_2^{(i)}+\dots+w_nx^{(i)}_n+b
$$

on 

$$
\textrm{suuri positiivinen, jos } y^{(i)}=1  $$
 $$\textrm{suuri negatiivinen, jos } y^{(i)}=0 
$$

eli 

$$
  a^{(i)}=\frac{1}{1+\exp(-z^{(i)})}
	\approx \left\{
	\begin{array}{l}
	1,\textrm{ jos } y^{(i)}=1\\
	0,\textrm{ jos } y^{(i)}=0\\
	\end{array}\right.
$$

 $$
\begin{array}{l}
\\\\
\end{array}
 $$

eli minimoidaan  kustannusfunktio (binary cross entropy)

$$
C=\displaystyle{\,-\,\frac{1}{m}  \,\sum_{i=1}^m y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})}$$ 

gradient descentillä.


Regularisoitu kustannusfunktio

$$
C=\displaystyle{\,-\,\frac{1}{m}  \,\sum_{i=1}^m y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})}+\frac{1}{2m}(w_1^2+w_2^2+...+w_n^2)$$ 


Pisteen $x=[x_1,x_2,\dots,x_n]$ luokittelu:

jos $z=w_1x_1+w_2x_2+\dots+w_nx_n+b> 0$, niin  luokka $y=1$ 




 Jos 
$z=w_1x_1+w_2x_2+\dots+w_nx_n+b< 0$
, niin  luokka $y=0$. 




Tulkinta: piste  kuuluu 
luokkaan 1 todennäköisyydellä 
$$a=\sigma(z)=\frac{1}{1+\exp(-z)}$$ 

ja luokaan 0 todennäköisyydellä $1-a$

## **2 muuttujaa, 3 luokkaa**









Datapisteet: $x^{(i)}=[x_1^{(i)},x_2^{(i)}],\quad      y^{(i)}=1,2,3, \quad  i=1,2,\dots m$

```python
k=50
m=3*k #datapisteiden määrä

P1=np.array([0.5,1])
P2=np.array([-1,0])
P3=np.array([0,-0.5])
r=0.25
xpun=r*np.random.randn(k,2)+P1
xvih=r*np.random.randn(k,2)+P2
xsin=r*np.random.randn(k,2)+P3
x=np.vstack((xpun,xvih,xsin))
ypun=np.ones(k)
yvih=2*np.ones(k)
ysin=3*np.ones(k)
y=np.hstack((ypun,yvih,ysin))

plt.figure(figsize=(5,5))
plt.plot(xpun[:,0],xpun[:,1],'r.',label='y=1',markersize=12)
plt.plot(xvih[:,0],xvih[:,1],'g.',label='y=2',markersize=12)
plt.plot(xsin[:,0],xsin[:,1],'b.',label='y=3',markersize=12)
plt.grid()
plt.legend(loc='best',fontsize=14)
plt.axis('equal')
plt.xlabel('$x_1$',fontsize=14)
plt.ylabel('$x_2 $',fontsize=14,rotation=0)
plt.show()
```

Etsitään  kertoimet 

$$w_{11},\dots,w_{23},\quad b_1,b_2,b_3$$

 niin, että piste $x=[x_1,x_2]$ kuuluu luokaan 1,2 tai 3 sen mukaan mikä luvuista

$$
\left\{
\begin{array}{l}
z_1=w_{11}x_1+w_{21}x_2+b_1\\
z_2=w_{12}x_1+w_{22}x_2+b_2\\
z_3=w_{13}x_1+w_{23}x_2+b_3
\end{array}
\right.
$$

on suurin

image.png

Matriiseilla (@ on matriisien kertolasku): 
$$[z_1,z_2,z_3]=[x_1,x_2]@\left[
\begin{array}{ccc}
w_{11}&w_{12}&w_{13}\\
w_{21}&w_{22}&w_{23}\\
\end{array}
\right]+[b_1,b_2,b_3]
$$

Tulkinta: piste $x$ kuuluu luokkaan $k=1,2,3$ todennäköisyydellä

$$
a_k=\frac{\exp(z_k)}{ \exp(z_1)+\exp(z_2)+\exp(z_3)}
$$



$$
\begin{array}{l}\\
\end{array}
$$



Lukuja $a_1,a_2,a_3$ sanotaan lukujen $z_1,z_2,z_3$ **softmax**iksi:



$$
a_1+a_2+a_3=1
$$




ja 
jos joku luvuista $z_k$ on paljon muita suurempi, niin 
$a_k\approx 1$ ja muut $a_j\approx 0$.


```python
#softmax
z=np.array([-1,-2,3,4])
a=np.exp(z)/np.sum(np.exp(z))
k=np.arange(1,len(z)+1)
plt.figure(figsize=(5,3))
plt.bar(k,a,width=0.4,zorder=2)#zorder=grid-viivat tolppien takana 
plt.grid()
plt.xlabel('k',fontsize=12)
plt.ylabel('ak  ',rotation=0,fontsize=12)
plt.xticks(k)
plt.title('z = ' + str(z))
plt.show()
```

Kertoimien $w_{jk}$ ja $b_k$ määrääminen: 
lasketaan  datapisteille $x^{(i)}=[x^{(i)}_1,x^{(i)}_2]$ 

$$\left\{\begin{array}{l}\\
\displaystyle{z^{(i)}_k=w_{1k}x^{(i)}_1+w_{2k}x^{(i)}_2+b_k}\\\\
\displaystyle{a^{(i)}_k=\frac{\exp(z^{(i)}_k)}{ \exp(z^{(i)}_1)+\exp(z^{(i)}_2)+\exp(z^{(i)}_3)}}
\end{array}\right.
,\quad k=1,2,3
$$

$$
\begin{array}{l}\\
\end{array}
$$


Jos datapisteen  luokka  $y^{(i)}=k$, niin sen kustannus 

$$
C^{(i)}=-\log(a_k^{(i)}),\quad k=1,2,3
$$

idea: 
$$
\begin{array}{l}
\textrm{jos   $a^{(i)}_k\approx 1$, niin  $C^{(i)}\approx 0$}\\\\   
\textrm{jos $a^{(i)}_k\approx 0$, niin    
$C^{(i)}$ on suuri}\\\\ %-\log(a_k^{(i)})\approx 0
\end{array}
$$


Minimoidaan kustannusfunktio
 (**categorical cross entropy**) 

$$
C=\frac{1}{m} \,\sum_{i=1}^m C^{(i)} $$ 

gradient descentillä.

Regularisoitu kustannusfunktio

$$
C=\frac{1}{m} \,\sum_{i=1}^m C^{(i)}+\frac{1}{2m}\sum_{j,k} (w_{jk})^2 $$ 


```python
from sklearn.linear_model import LogisticRegression
#datapisteiden koordinaatit [x1,x2] matriisin x vaakariveinä
#luokat 1,2,3 1D-vektorissa y
model=LogisticRegression().fit(x,y) #minimoi regularisoidun kustannusfunktion
#model=LogisticRegression(penalty=None).fit(x,y) #minimoi regularisoimattoman kustannusfunktion
 
w=model.coef_.T  #HUOM: sklearnin antama kerroinmatriisi on ylläolevan transpoosi !
b=model.intercept_

np.set_printoptions(precision=3)
print('w = '+str(w))
print('\n')

print('b = '+str(b))
print('\n')

ypred=model.predict(x) #datapisteiden ennustetut luokat 1,2,3

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y, ypred)) #rivi = luokka, sarake = ennuste
```

```python
#testipiste
xt=np.array([[1,1]])
yt=model.predict(xt) #luokka
print(yt)
at=model.predict_proba(xt) #softmax
print(at)
print('\n')

#sama kertoimilla
z=xt@w+b #vaakavektori [z1,z2,z3]
print(z)
a=np.exp(z)/np.sum(np.exp(z)) #softmax
print(a)
```

**Huom:**  Luokkien reunat ovat suoria

Luokan 1  pisteet  $x=[x_1,x_2]$ toteuttavat ehdot




$$\begin{array}{rl}
z_1>z_2 &
\leftrightarrow  w_{11}x_1+w_{21}x_2+b_1> w_{12}x_1+w_{22}x_2+b_2\\\\
&\leftrightarrow(w_{11}-w_{12})x_1+(w_{21}-w_{22})x_2+(b_1-b_2)> 0
\end{array}
$$



$$\begin{array}{rl}
&\\\\z_1>z_3 &
\leftrightarrow  w_{11}x_1+w_{21}x_2+b_1> w_{13}x_1+w_{23}x_2+b_3\\\\
&\leftrightarrow(w_{11}-w_{13})x_1+(w_{21}-w_{23})x_2+(b_1-b_3)> 0
\\\\&\end{array}
$$

eli luokan 1 reunan muodostavat vektoreita 

$$[w_{11}-w_{12},w_{21}-w_{22}]\quad \textrm{ja}\quad [w_{11}-w_{13},w_{21}-w_{23}]$$

vastaan kohtisuorat suorat 

$$
(w_{11}-w_{12})x_1+(w_{21}-w_{22})x_2+(b_1-b_2)= 0
$$
ja 
$$
(w_{11}-w_{13})x_1+(w_{21}-w_{23})x_2+(b_1-b_3)= 0
$$

Luokat 2 ja 3 vastaavasti.

\end{array}

\end{array}

\end{array}
