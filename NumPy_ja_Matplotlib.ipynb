##**NumPy**

```python
import numpy as np

#nD-array

#1D-array = 1D-vektori
#2D-array = 2D-matriisi eli kokoelma 1D-arrayita
#3D-array = 3D-matriisi eli kokoelma 2D-arrayita, esim rgb-kuva
#4D-array = 4D-matriisi eli kokoelma 3D-arrayita, esim rgb-kuvia
```

```python
#1D-array = vektori
v=np.array([4,5,1,0,-2])  #v=[v[0],v[1],v[2],v[3],v[4]]
#HUOM: indeksointi alkaa 0:sta !!
v
```

```python
v.shape #  v:n muoto (koko) = (5,)
v[0] #alkio 0
v[0:3] #alkiot 0,1,2 HUOM: alkio v[3] EI mukana !
v[1:-1] #alkiot 1,2,...,viimeistä edellinen
v[1:] #alkiot 1,2,...
v[-3:] #alkiot a[2],a[3],a[4] eli kolmanneksi viimeinen ... viimeinen
ind=[1,2,4]
v[ind] #alkiot 1,2 ja 4
```

```python
#tasavälisiä vektoreita
v=np.arange(0,10,2) #(alku,loppu,askel), loppu EI mukana
print(v,'\n')
w=np.linspace(0,1,11) #(alku,loppu,lukumäärä), loppu ON mukana
print(w)
```

```python
#2D-array = matriisi
A=np.array([[1,3,4,0],[-1,5,8,6],[5,3,0,1]]) 
#vaakarivit [1,3,4,0],[-1,5,8,6],[5,3,0,1]
A
```

```python
m,n=A.shape #a:n koko
#m=vaakavien lkm
#n=pystyrivien lkm
r=0
s=1
A[r,s] #alkio vaakarivillä r, pystyrivillä s

A[r,:] #tai A[r], vaakarivi r, 1D-vektori !!
A[:,s] #pystyrivi s, 1D-vektori !!

A[1:3,0:3] #vaakarivit 1 ja 2, pystyrivit 0,1 ja 2
A[[0,2],:] #vaakarivit 0 ja 2
A[:,[0,2,3]] #pystyrivit 0,2,3 
```

```python
#satunnaislukuja
m=2
n=3
#tasaisesti jakautuneita lukuja väliltä 0...1
np.random.rand(m,n) #(m,n)-matriisi
np.random.rand(m) #(m,)-vektori
np.random.rand() #1 luku

#tasaisesti jakautunut satunnaisluku väliltä a...b
a=3
b=5
a+(b-a)*np.random.rand() 

#tasaisesti jakautuneita kokonaislukuja väliltä xmin...xmax-1
xmin=-4
xmax=5
np.random.randint(xmin,xmax,(m,n))  #(m,n)-matriisi
np.random.randint(xmin,xmax,(m)) #(m,)-vektori 
np.random.randint(xmin,xmax) #1 luku 

#normaalijakautuneita lukuja, keskiarvo = 0, keskihajonta = 1
np.random.randn(m,n) #(m,n)-matriisi
np.random.randn(m) #(m,)-vektori
np.random.randn() #1 luku

#normaalijakautunut satunnaisluku, keskiarvo = mu, keskihajonta = sigma
mu=3
sigma=2
mu+sigma*np.random.randn() 
```

```python
#vektorin/matriisin luonti isommista paloista

a=np.array([1,2,3])
b=np.array([4,5,6])

abh=np.hstack((a,b)) #horizontally eli vierekkäin
print(abh,'\n')

abv=np.vstack((a,b)) #vertically eli päällekkäin 
print(abv,'\n')

#a ja b matriisin abr riveiksi 
abr=np.zeros((2,3))
abr[0,:]=a
abr[1,:]=b
print(abr,'\n')

#a ja b matriisin abs sarakkeiksi
abs=np.zeros((3,2))
abs[:,0]=a
abs[:,1]=b
print(abs,'\n')
```

```python
#max, min, sum, mean, std
m=4
n=5
A=np.random.randint(-4,4,(m,n))

np.max(A) #alkioiden maksimiarvo
np.min(A) #alskioiden minimiarvo
np.max(A,axis=0) #sarakkeiden maksimit, 1D-vektori, (n,)
np.max(A,axis=0,keepdims=True) #sarakkeiden maksimit, vaakavektori, (1,n)
np.max(A,axis=1) #rivien maksimit, 1D-vektori, (m,)
np.max(A,axis=1, keepdims=True) #rivien maksimit, pystyvektori, (m,1)

np.sum(A)  #alkioiden summa
np.sum(A,axis=0) #sarakkeiden summat
np.sum(A,axis=0,keepdims=True)
np.sum(A,axis=1) #rivien summat
np.sum(A,axis=1,keepdims=True)

np.mean(A) #alkioiden keskiarvo
np.mean(A,axis=0) #sarakkeiden keskiarvot
np.mean(A,axis=0,keepdims=True) 
np.mean(A,axis=1) #rivien keskiarvot
np.mean(A,axis=1,keepdims=True)

np.std(A) #alkioiden keskihajonta
np.std(A,axis=0) #sarakkeiden keskihajonnat
np.std(A,axis=0,keepdims=True) 
np.std(A,axis=1) #rivien keskihajonnat
np.std(A,axis=1,keepdims=True) 
```

```python
#Boolean masking eli indeksointi true/false-ehdoilla
A=np.random.randint(-5,5,(4,5))
print(A,'\n')

print(A[A>0],'\n') #alkiot, jotka >0, 1D-vektori

y=np.array([1,0,0,1,1])
print(y==1,'\n')
print(A[:,y==1],'\n') #sarakkeet 0,3,4

y=np.array([0,1,0,1])
print(A[y==1,:]) #rivit 1,3

#and: &
#or:  | 
#not:  ~
#not equal:  !=
```

matriisi + samankorkuinen pystyvektori: yhteenlasku sarakkeittain

```python
#(m,n)+(1,n) = (m,n), rivit+vaakavektori
#(m,n)-(1,n) = (m,n), rivit-vaakavektori
#(m,n)*(1,n) = (m,n), rivit*vaakavektori
#(m,n)/(1,n) = (m,n), rivit/vaakavektori
#(m,n)+(m,1) = (m,n), sarakkeet+pystyvektori
#(m,n)-(m,1) = (m,n), sarakkeet-pystyvektori
#(m,n)*(m,1) = (m,n), sarakkeet*pystyvektori
#(m,n)/(m,1) = (m,n), sarakkeet/pystyvektori
#(1,n)+(m,1) = (m,n), alkio ij = (1,i)+(j,1)
#(1,n)-(m,1) = (m,n), alkio ij = (1,i)-(j,1)
#(1,n)*(m,1) = (m,n), alkio ij = (1,i)*(j,1)
#(1,n)/(m,1) = (m,n), alkio ij = (1,i)/(j,1)


#HUOM: broadcastingissa 1D-vektori =  vaakavektori
#(m,n)+(n,) =(m,n), rivit + 1D-vektori
```

```python
a=np.array([1,2,3]) #(3,)
b=np.array([[4,5,6]]) #(1,3)
c=np.array([[8],[9],[10]]) #(3,1)
d=np.ones((3,3))
e=np.ones((4,3)) 
f=np.array([1,2,3,4]) #(4,)
g=f.reshape((4,1)) #(4,1)

a+b #(3,)+(1,3)=(1,3) #alkioittain
b+c #(1,3)+(3,1)=(3,3) #alkio ij = b[i]+c[j]
d+a #(3,3)+(3,)=(3,3), d:n vaakarivit+a
d+b #(3,3)+(1,3)=(3,3), d:n vaakarivit+a
d+c #(3,3)+(3,1)=(3,3), d:n pystyrivit+c
a+e #(3,)+(4,3)=(4,3), a+e:n vaakarivit
b+e #(1,3)+(4,3)=(4,3), b+e:n vaakarivit
#c+e #(3,1)+(4,3) ei käy, ValueError: operands could not be broadcast together with shapes (3,1) (4,3) 
#e+f #(4,3)+(4,) ei käy
e+g #(4,3)+(4,1)=(4,3), e:n pystyrivit + g
```

```python
#esim: datan skaalaus
m=6
n=5
X=np.random.rand(m,n)
print(X,'\n')
#"datapisteet" X:n vaakariveinä
#X:n sarakkeiden eli "koordinaattien" keskiarvot
mu=np.mean(X,axis=0)
print(mu,'\n')
#ja keskihajonnat
sigma=np.std(X,axis=0)
print(sigma,'\n')
#matriisin 
Xhat=(X-mu)/sigma 
print(Xhat,'\n')
#sarakkeiden keskiarvot ovat = 0 ja -hajonnat = 1
print(np.mean(Xhat,axis=0),'\n')
print(np.std(Xhat,axis=0))
```

```python
#transpoosi 
A=np.array([[1,2,3],[4,5,6]])
print(A.T,'\n')

np.array([1,2,3]).T #1D-vektoreille ei vaikutusta
```

```python
#matriisikertolasku 2D-arrayille 
A=np.array([[1,2],[4,7]])
B=np.array([[3,5],[-4,2]])
print(A@B,'\n')

#käänteismatriisi
A_inv=np.linalg.inv(A)
print(A_inv)
```

##**Matplotlib**

```python
import matplotlib.pyplot as plt

# Jos x ja y ovat yhtä pitkiä 1D-vektoreita (tai Python-listoja [a,b,c,...]),
# niin plt.plot(x,y)
# yhdistää pisteet [x[0],y[0]] , [x[1],y[1]], [x[2],y[2]], … viivalla
# eli  vaaka-akselilla x, pystyakselilla y
# plt.plot(y) -> vaaka-akselilla järjestysnumero 0,1,2,..., pystyakselilla y

n=11
x=np.linspace(-1,1,n)
y1=np.random.rand(n)
y2=x**2

plt.figure(figsize=(8,6)) #(leveys,korkeus)
plt.plot(x,y1,'r',linewidth=2,label='y1') #punainen viiva
plt.plot(x,y2,'g-o',markersize=10,label='y2') #vihreä viiva + ympyrä
plt.plot([-1,0,1],[0.5,-0.4,1.4],'k',label='viiva') #musta viiva
plt.plot(x[0],y1[0],'ms',label='piste') #magenta neliö
plt.title('otsikko',fontsize=18)
plt.xlabel('vaaka-akselin otsikko',fontsize=18)
plt.ylabel('pysty-akselin otsikko')
plt.grid() #taustaristikko
plt.legend(fontsize=18) #värikartta
plt.xlim(-1.1,1.1) #vaaka-akselin rajat
plt.xticks(np.linspace(-1,1,11)) #x-akselin apuviivat
plt.ylim(-1,2) #pystyakselin rajat
plt.show()
```

```python
#esim: luetaan scikit-learn-kirjaston iris-datajoukko
from sklearn import datasets
X,y=datasets.load_iris(return_X_y=True)
#matriisin X rivit datapisteitä [x0,x1,x2,x3]
#datapisteiden luokat y=0,1,2 1D-vektorissa y
print(X.shape,'\n')
print(y.shape,'\n')
print(X[0,:],'\n')
print(y[0])
```

```python
#kuva datapisteistä käyttäen koordinaatteja (eli X:n sarakkeita) i ja j
i=0 
j=1

plt.figure(figsize=(5,5))
plt.plot(X[y==0,i],X[y==0,j],'r.',label='y=0') #luokan y=0 pisteet 
plt.plot(X[y==1,i],X[y==1,j],'g.',label='y=1') #luokan y=1 pisteet
plt.plot(X[y==2,i],X[y==2,j],'b.',label='y=2') #luokan y=2 pisteet
plt.grid()
plt.legend()
plt.xlabel('xi')
plt.ylabel('xj',rotation=0)
plt.show()
```

```python
#skaalataan koordinaatit: keskiarvoiksi 0 ja -hajoinnoiksi 1
mu=np.mean(X,axis=0)
sigma=np.std(X,axis=0)
Xhat=(X-mu)/sigma
```

```python
#alkuperäisen ja skaalatun koordinaatin xj arvot 
j=0
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X[:,j],'.')
plt.grid()
plt.title('alkuperäinen')
plt.xlabel('datapiste numero')
plt.subplot(122)
plt.plot(Xhat[:,j],'.')
plt.grid()
plt.title('skaalattu')
plt.xlabel('datapiste numero')
plt.show()
```

```python
#alkuperäisen ja skaalatun koordinaatin xj arvojen jakaumat
j=0
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(X[:,j],bins=20,edgecolor='k',zorder=2)
plt.xlabel('xj')
plt.grid()
plt.title('alkuperäinen')
plt.subplot(122)
plt.hist(Xhat[:,j],bins=20,edgecolor='k',zorder=2)
plt.xlabel('xj')
plt.grid()
plt.title('skaalattu')
plt.show()
```
