## **HARJOITUS 2:**



Etsi allaolevien  datapisteiden $x^{(i)},y^{(i)}$ perusteella suureelle $y$ laskukaava

$$
y=w_1x^2+w_2x+b
$$

käyttämällä lineaarista regressiota datapisteille 

$$
[x_1^{(i)},x_2^{(i)}],\,y^{(i)}
$$

missä 

$$x_1^{(i)}=(x^{(i)})^2,\,x_2^{(i)}=x^{(i)}$$

kuva.png

ja piirrä löytämäsi paraabelin kuvaaja samaan kuvaan datapisteiden kanssa.

kuva.png

```python
#datapisteet (paraabelin y=Ax^2+Bx+C pisteet + virhe)
m=30 
A=1
B=2
C=-4
x0=-B/(2*A) #paraabelin huippu
L=3 
np.random.seed(0)

x=x0-L+2*L*np.random.rand(m,1) #väliltä x0-L...x0+L
virhe=0.4*np.random.randn(m,1)
y=A*x**2+B*x+C+virhe

```
